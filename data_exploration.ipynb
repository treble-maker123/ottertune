# Data Exploration

```python
import pandas as pd
from IPython.display import display

def print_sample(file_path: str, print_headers: bool = False):
    df = pd.read_csv(file_path)
    metrics = df.columns.values.tolist()[1:] # first one is "workload id"
    
    if print_headers:
        for header in metrics:
            print(header)

    display(df)
```

## offline_workload.csv

```python
offline_workload_path = 'project3_dataset/offline_workload.CSV'
print_sample(offline_workload_path)
```

## online_workload_B.CSV

```python
online_workload_b_path = 'project3_dataset/online_workload_B.CSV'
print_sample(online_workload_b_path)
```

## online_workload_C.csv

```python
online_workload_c_path = 'project3_dataset/online_workload_C.CSV'
print_sample(online_workload_c_path)
```

## test.csv

```python
test_path = 'project3_dataset/test.CSV'
print_sample(test_path)
```
